## The power of asking questions:

1. Sergey Brin and Larry Page changed the world by asking themselves a simple question: <br />**How can we use data to rank the importance of websites?**<br /> This was a simple yet valuable question because it led to the invention of the famous **PageRank** [algorithm](https://www.mathsisfun.com/definitions/algorithm.html). **PageRank** is the technology that Google uses to recommend websites when people do Google searches and it helped to create the trillion dollar tech empire that Google is today.

Once you have your topic, it is time to ask yourself some questions and discuss with your teammates. For example, if your topic is to identify symptoms of climate change in your local area, you can ask the following questions:
- Have I noticed any changes in the climate since my childhood? (e.g. The beach I used to go to disappeared; I experienced the hottest summer last year in Seattle; We expericed the earliest cherry blossom in D.C last year.)
- What are the kinds of indicators (data) we can look at? (e.g., temperature, precipation, ocean heat, humidity etc.)
- Where can we find these data? (e.g. [National Centers for Environmental Inofrmation](https://www.ncdc.noaa.gov/cdo-web/), [Wildlife Dataset](https://www.systemanaturae.org/datasets/), etc)

3. Record ALL of the questions that come into your mind

# <div align="center">  Activity (Optional): **Practice With Your Own Mini "Dear Data" Visual!**

***
Get ready! It's time for you to practice asking questions by working on your own "Dear Data activity. Woohoo!

<div align="center"> 
<figure class="image">
  <img src="http://datastori.es/wp-content/uploads/2015/11/Stefanie_DearData_14-front.jpg", width = '500'>
  <figcaption> Dear Data Visual <br \> Source: dear-data.com </figcaption>
</figure>
<div align="left">

The goal of this activity is for you to practice asking questions that you can answer using data about your own life! Here's how it'll work.

**Step 1**: Pick an activity or behavior that you do regularly. This might include watching TV, laughing, doing your homework, listening to music, spending time on social media, saying "thank you" to your parents, cleaning your room, or anything else that you do.

**Step 2**: **Ask a question** about that activity or behavior! For example, if you laugh a lot, you could ask, "What are the kinds of things that make me laugh the most?" If you watch a lot of TV, you might ask, "What kinds of TV shows do I spend the most time watching?"

**Step 3**: Figure out how to record or collect your data for the next **4-5 days**! This doesn't need to be fancy. We're going to talk more about collecting data next week. For now, you just need to think about how you'll track the information. Make sure to keep it all in one place and try to collect the same kind of information every time. Using the laughing example from Step 2, you might decide that every time you laugh, you'll right down the the specific thing that made you laugh (a TV show? a friend? a pet?) and where you were at the time.

**Step 4**: This is the fun part!! After you collect your data, turn them into a visual! Don't worry, you don't need to be an artist and your visual doesn't need to be a work of art. It's perfectly okay for you to hand-draw it. The main focus here is telling a story with the data that you collected - there is no right or wrong answer because you are telling the story the way you want to! [Here are some Dear Data Tips to help you with your creation!](https://drive.google.com/file/d/193HWX7myADflIK1XAAgQ0zVegw1dyQBX/view?usp=sharing)

**Step 5**: Take a picture of your Dear Data visual and upload it using this [Google Form!](https://docs.google.com/forms/d/e/1FAIpQLSdOGX_NBg3elpXPrTtTf8lvi_JTx28sUxInxhqbIBbP0vcNrw/viewform) We'd love to see your work!!
