# <div align="center">  New Information to Explore: **Data Visualization**

## Why is data visualization important?

Good visualization makes the data more natural for the human mind to comprehend and therefore makes it easier to identify trends, patterns, and outliers within large data sets. Visualization is more important when we have multivariate data.

## Example:
